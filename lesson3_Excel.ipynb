```python
import pandas as pd
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
export_2016 = pd.read_excel('WITS_Russia_export_2016.xlsx', sheet_name='Partner')
export_2017 = pd.read_excel('WITS_Russia_export_2017.xlsx', sheet_name='Partner')
```

```python
export_2016.index
```

```python
export = pd.concat([export_2016, export_2017]) 
```

```python
export.sample(10)
```

```python
export.pivot_table(index=['Year'], values='Export (US$ Thousand)', aggfunc='sum')
```

```python
export.pivot_table(index=['Year', 'Partner Name'], values='Export (US$ Thousand)', aggfunc='sum')
```

```python
export.pivot_table(index=['Year', 'Partner Name'], values='Export (US$ Thousand)', aggfunc='sum')\
.sort_values('Export (US$ Thousand)', ascending=False)
```

```python
export_filtered = export[export['Partner Name'] != 'Europe & Central Asia']
export_filtered.pivot_table(index=['Year', 'Partner Name'],
    values='Export (US$ Thousand)', aggfunc='sum')\
    .sort_values('Export (US$ Thousand)', ascending=False)
```

```python
exclude = ['East Asia & Pacific', 'Europe & Central Asia', ' World', 'Unspecified', 'Middle East & North Africa']
export_filtered = export[-export['Partner Name'].isin(exclude)]
export_filtered.pivot_table(index=['Year', 'Partner Name'],
    values='Export (US$ Thousand)', aggfunc='sum')\
    .sort_values('Export (US$ Thousand)', ascending=False)
```

```python
export_pivot = export_filtered.pivot_table(index=['Year', 'Partner Name'], 
                values='Export (US$ Thousand)', aggfunc='sum')\
            .sort_values(['Year', 'Export (US$ Thousand)'],ascending=[True, False])
```
